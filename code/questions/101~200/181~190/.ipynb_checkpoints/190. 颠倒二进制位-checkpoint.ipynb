[190. 颠倒二进制位](https://leetcode-cn.com/problems/reverse-bits/)

```python
class Solution:
    # @param n, an integer
    # @return an integer
    def reverseBits(self, n):
        rst = 0
        for _ in range(32):
            rst = (rst << 1) + (n & 1)
            n >>= 1
        return rst

Solution().reverseBits(5)
```

```python
Solution().reverseBits(2)
```
