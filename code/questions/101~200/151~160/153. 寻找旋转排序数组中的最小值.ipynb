[153. 寻找旋转排序数组中的最小值](https://leetcode-cn.com/problems/find-minimum-in-rotated-sorted-array/)

```python
class Solution(object):
    def findMin(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        i, j = 0, len(nums)-1
        while i < j:
            mid = i + (j-i)//2
            if nums[i] > nums[mid]:
                j = mid
            elif nums[mid] > nums[j]:
                i = mid + 1
            else:
                j = i
        return nums[i]
```
