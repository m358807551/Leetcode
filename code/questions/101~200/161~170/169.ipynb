[169. 多数元素](https://leetcode-cn.com/problems/majority-element/)

```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        rst, life = None, 0
        for num in nums:
            if rst is None:
                rst = num
                life = 1
            elif rst == num:
                life += 1
            elif rst != num:
                life -= 1
                if life == 0:
                    rst = None
        return rst
```

```python
Solution().majorityElement([2, 1, 1, 1, 2, 2, 2])
```
