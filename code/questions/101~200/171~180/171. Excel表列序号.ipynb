[171. Excel表列序号](https://leetcode-cn.com/problems/excel-sheet-column-number/)

```python
class Solution(object):
    def titleToNumber(self, s):
        """
        :type s: str
        :rtype: int
        """
        rst = 0
        for letter in s:
            rst = rst * 26 + ord(letter) - 64
        return rst
```

```python
Solution().titleToNumber("AB")
```
