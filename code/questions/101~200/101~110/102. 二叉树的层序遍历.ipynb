[102. 二叉树的层序遍历](https://leetcode-cn.com/problems/binary-tree-level-order-traversal/)

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def levelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        rst = []
        queue = [(root,)]
        while queue:
            line = [ 
                node.val
                for nodes in queue
                for node in nodes
                if node
            ]
            if line:
                rst.append(line)
            queue = [
                (node.left, node.right)
                for nodes in queue
                for node in nodes
                if node
            ]
        return rst
```
