```python
[78. 子集](https://leetcode-cn.com/problems/subsets/)
```

```python
class Solution(object):
    def subsets(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        rst = [[]]
        for num in nums:
            rst.extend([x+[num] for x in rst])
        return rst
    
```

```python
Solution().subsets([1, 2, 3])
```
